```python
import skimage.io
import numpy as np
```

```python
img = skimage.io.imread("Academic_day2/GFP.tif")
img = img[0]
skimage.io.imshow(img)
```

```python
print("Max value is", np.amax(img))
# define the threshold as a 10% of maximum intensity to avoid background pixels
thresh = np.amax(img) *0.1
print("Background value is", thresh)
```

Make binary image in the range of values between 0 and 1

```python
new_img = np.where(img > thresh, 1, 0)
skimage.io.imshow(new_img)
```

```python
# Calculate the mean of pixels equal 1
print("The mean value of a cell is", np.mean(img[new_img== 1]))
```

```python
img_3d = skimage.io.imread("Academic_day2/GFP.tif")
print(np.shape(img_3d))
mean_intensity = []
for i, frame in enumerate(img_3d):
  thresh = np.amax(frame) *0.1
  BW_image = np.where(frame > thresh, 1, 0)
  mean_intensity.append(np.mean(frame[BW_image== 1]))
  print(f"{i} {mean_intensity[i]:.3f}")
```

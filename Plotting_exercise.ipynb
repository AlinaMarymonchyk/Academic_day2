```python
def handle_significance(p_value: int):
    if p_value < 0.001:
        return "***"
    if p_value < 0.01:
        return "**"
    if p_value < 0.05:
        return "*"
    else:
        return False
```

```python
def multi_box_plot_analysis(df: pd.DataFrame,
                            columns: str,
                            populations: str,
                            gridsize=(2, 2)):

    legendBool = True
    fig = make_subplots(gridsize[0], gridsize[1], subplot_titles=columns)
    m_comp_df = pd.DataFrame(columns=["Parameter", "Group1", "Group2", "p-value", "Significant"])
    fig.update_layout(template="plotly_white")
    p_value_df = None
    colorsList = ["blue", "green", "red"]
    for graphNumber, col in enumerate(columns):
        dataGroup = {}
        for j, pop_ in enumerate(df[populations].unique()):

            df2 = df[df[populations] == pop_]

            pop_color = j

            if graphNumber > 0:
                legendBool = False
                # To avoid to get multiple legends we set it only for first graph
            fig.add_trace(go.Box(
                y=df2[col],
                name=f"{pop_}",
                boxmean = True,
                fillcolor="rgba(0,0,0,0)",
                pointpos=0,boxpoints="all",
                marker=dict(size=1, ),
                line=dict(width=1),
                marker_color=px.colors.qualitative.Plotly[pop_color],
                opacity=1,
                legendgroup=pop_,
                showlegend=legendBool),
                row=graphNumber // gridsize[1] + 1,
                col=graphNumber % gridsize[1] + 1)

        for a, b in itertools.combinations(df[populations].unique(), 2):
            m_comp = stats.ttest_ind(df[df[populations] == a][col], df[df[populations] == b][col])
            m_comp_df = pd.concat([m_comp_df, pd.DataFrame([[col, a, b, m_comp.pvalue, handle_significance(m_comp.pvalue)]],
                                          columns=["Parameter", "Group1", "Group2", "p-value", "Significant"])])

    return m_comp_df, fig
```

```python
parameters = ["Parameter 31",
     "Parameter 33",
     "Parameter 1",
     "Parameter 2"]
```

```python
multi_box_plot_analysis(example, parameters, populations="Populations", gridsize=(2, 2))[1]
```

```python
multi_box_plot_analysis(example, parameters, populations="Populations", gridsize=(2, 2))[0]
```
